1 Tuplas (Imutável)

```python
nomes = ('Goku','Vegeta','Trunks','Gohan')
print(nomes)
#SELECT(slicing de Dados)
print(nomes[0])
print(nomes[2:])
print(nomes[1:3])
print(nomes[-1])
```

```python
#DELETE
del nomes[3]
print(nomes)
```

2 Listas (Mutável)


```python
nomes  = ['Goku','Vegeta','Trunks','Gohan']
```

```python
#SELECT
#funciona exatamente como a tupla
```

```python
#INSERT
nomes.append('Bulma')#Insert no final
nomes.insert(1,'Piccolo')#Insere na posição indicada
print(nomes)
```

```python
#UPDATE
nomes[0] = 'Kami'
print(nomes)
```

```python
#DELETE
del nomes[0] # Exclui elementos pelo indice
#nomes.remove('Trunks') # Exclui elementos pelo valor
print(nomes)
```

3 Conjuntos set (elementos únicos)

```python
nomes  = {'Goku','Vegeta','Trunks','Gohan','Goku'}
#SELECT
print(nomes)
```

```python
#DELETE
nomes.remove('Gohan')
print(nomes)
```

```python
#UPDATE
#DELETE UM ELEMENTO E DEPOIS ADICIONA OUTRO
```

```python
#INSERT
nomes.add('Piccolo')
print(nomes)
```

4 Dicionário (mutável//key&value)

```python

#INSERT
pessoa = {'nome': 'Goku',
          'idade': 30,
          'altura': 1.80}

print(pessoa)

#SELECT
print(pessoa['nome'])
#UPDATE
pessoa['idade'] = 42
#DELETE
del pessoa['altura']
#INSERT
pessoa['sexo'] = 'Masculino'
print(pessoa)
```
